```python
import asyncio
from playwright.async_api import async_playwright
import os

async def save_url_as_pdf_colab(url: str):
    """
    Truy c·∫≠p m·ªôt URL v√† l∆∞u n·ªôi dung trang web d∆∞·ªõi d·∫°ng PDF.
    T√™n file PDF s·∫Ω d·ª±a tr√™n ti√™u ƒë·ªÅ trang (title).

    Args:
        url (str): URL c·ªßa trang web.
    """
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        try:
            print(f"ƒêang truy c·∫≠p: {url}...")
            await page.goto(url, wait_until='networkidle')

            # L·∫•y ti√™u ƒë·ªÅ trang ƒë·ªÉ ƒë·∫∑t t√™n file
            title = await page.title()
            safe_title = "".join(c for c in title if c.isalnum() or c in (' ', '_', '-')).rstrip()
            filename = safe_title.replace(" ", "_") + ".pdf"
            output_path = os.path.join("/content/", filename)

            print(f"ƒêang l∆∞u PDF: {filename}...")
            await page.pdf(path=output_path, format='A4', print_background=True)

            print(f"‚úÖ ƒê√£ l∆∞u: {output_path}")

        except Exception as e:
            print(f"‚ùå L·ªói v·ªõi {url}: {e}")
        finally:
            await browser.close()

async def main():
    urls = [
        "https://vi.wikipedia.org/wiki/ƒê·∫°i_h·ªçc_Qu·ªëc_gia_H√†_N·ªôi",
        "https://vi.wikipedia.org/wiki/Tr∆∞·ªùng_ƒê·∫°i_h·ªçc_C√¥ng_ngh·ªá,_ƒê·∫°i_h·ªçc_Qu·ªëc_gia_H√†_N·ªôi",
        "https://uet.vnu.edu.vn/lich-su/",
        "https://handbook.uet.vnu.edu.vn/H%E1%BB%8Dc%20ph%C3%AD%20-%20Ch%E1%BA%BF%20%C4%91%E1%BB%99%20ch%C3%ADnh%20s%C3%A1ch/",
        "https://handbook.uet.vnu.edu.vn/H%E1%BB%8Dc%20b%E1%BB%95ng/",
        "https://uet.vnu.edu.vn/vien-tri-tue-nhan-tao/"
    ]

    for url in urls:
        await save_url_as_pdf_colab(url)

# Ch·∫°y trong Colab ho·∫∑c Jupyter
await main()
```

```python
import os
from google import genai
from google.genai import types

def generate_markdown_from_pdfs(folder_path="/content"):
    client = genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))
    model = "gemini-2.5-flash-preview-04-17"

    # L·∫∑p qua t·∫•t c·∫£ file PDF trong th∆∞ m·ª•c
    pdf_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf")]

    if not pdf_files:
        print("‚ùå Kh√¥ng t√¨m th·∫•y file PDF n√†o trong th∆∞ m·ª•c.")
        return

    for pdf_file in pdf_files:
        full_pdf_path = os.path.join(folder_path, pdf_file)
        print(f"\nüìÑ ƒêang x·ª≠ l√Ω: {pdf_file}")

        try:
            uploaded_file = client.files.upload(file=full_pdf_path)
        except Exception as e:
            print(f"‚ùå L·ªói upload file {pdf_file}: {e}")
            continue

        contents = [
            types.Content(
                role="user",
                parts=[
                    types.Part.from_uri(
                        file_uri=uploaded_file.uri,
                        mime_type=uploaded_file.mime_type,
                    ),
                    types.Part.from_text(text=(
                        "Chuy·ªÉn to√†n b·ªô file PDF sang ng√¥n ng·ªØ Markdown, "
                        "l∆∞u √Ω chuy·ªÉn ƒë·ªïi ch√≠nh x√°c c√°c c√¥ng th·ª©c, b·∫£ng bi·ªÉu. "
                        "Output ch·ªâ ch·ª©a n·ªôi dung Markdown."
                    ))
                ],
            )
        ]

        generate_content_config = types.GenerateContentConfig(
            response_mime_type="text/plain"
        )

        output_lines = []

        try:
            for chunk in client.models.generate_content_stream(
                model=model,
                contents=contents,
                config=generate_content_config,
            ):
                output_lines.append(chunk.text)
        except Exception as e:
            print(f"‚ùå L·ªói sinh n·ªôi dung t·ª´ {pdf_file}: {e}")
            continue

        # Ghi k·∫øt qu·∫£ ra file .txt v·ªõi t√™n t∆∞∆°ng ·ª©ng
        output_txt_path = os.path.join(
            folder_path, os.path.splitext(pdf_file)[0] + ".txt"
        )
        try:
            with open(output_txt_path, "w", encoding="utf-8") as f:
                f.write("".join(output_lines))
            print(f"‚úÖ ƒê√£ l∆∞u Markdown v√†o: {output_txt_path}")
        except Exception as e:
            print(f"‚ùå L·ªói khi ghi file: {output_txt_path} - {e}")

if __name__ == "__main__":
    generate_markdown_from_pdfs("/content")
```
